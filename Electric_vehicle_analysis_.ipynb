```python
path = ("/content/Electric_Vehicle_Population_Data_Excel_sheet.xlsx")
ev_data = pd.read_excel(path)
print(ev_data.head())
```

```python
#ev adpotion over time - analyze growth of population by model year

import matplotlib.pyplot as plt
import seaborn as sns

sns.style=('whitegrid')

plt.figure(figsize =(12,6))

ev_adoption_by_year = ev_data['Model Year'].value_counts().sort_index()

print(ev_adoption_by_year)

sns.barplot(x= ev_adoption_by_year.index,
            y= ev_adoption_by_year.values,
            palette = "viridis"
            )

plt.title("Electrice vehicle adoption over time")
plt.xlabel("Model Year")
plt.ylabel("Number of vehicle registered")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# geographical distrbution at county level - selecting top 3 counties based  ev



ev_county_distribution = ev_data['County'].value_counts()

print(ev_county_distribution)

top_counties = ev_county_distribution.head(3).index

print(top_counties)

#filter  dataset for top counties

top_counties_data = ev_data[ev_data['County'].isin(top_counties)]

print(top_counties_data)



#analyze the distribution of ev within cities fo these counties



print(ev_county_distribution)

top_counties = ev_county_distribution.head(3).index

print(top_counties)

#filter  dataset for top counties


top_counties_data = ev_data[ev_data['County'].isin(top_counties)]

print(top_counties_data)



#analyze the distribution of ev within cities fo these counties

ev_city_distribution_top_counties = top_counties_data.groupby(['County','City']).size().sort_values(ascending = False)

print(ev_city_distribution_top_counties)

top_cities = ev_city_distribution_top_counties.head(10)

print(top_cities)

plt.figure(figsize= (12,6))


sns.barplot(
    x='City',
    y='number of vehicles',
    hue ='County',
    data = top_cities.reset_index().rename(columns={0: 'number of vehicles'}),
    palette ="magma"
)

plt.title("geographical Distribution of electric vehicle(top 10 cities)")
plt.xlabel("city")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#types of electric vehicle

ev_type_distribution = ev_data['Electric Vehicle Type'].value_counts()

print(ev_type_distribution)

plt.figure(figsize = (12,6))

sns.barplot(
    x=ev_type_distribution.index,
    y=ev_type_distribution.values,
    palette = "viridis"

)

plt.title("Types of Electric Vehicle")
plt.xlabel("Electric Vehicle Type")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
#popularity of electric vehicle manufacturer and model among registered vehicle

ev_maufacturer_distribution = ev_data['Make'].value_counts()

print(ev_maufacturer_distribution.head(10))

plt.figure(figsize = (12,6))

sns.barplot(
    x= ev_maufacturer_distribution.head(10).index,
    y= ev_maufacturer_distribution.head(10).values,
    palette = "cubehelix"

)

plt.title("Popularity of electric vehicle maufacturer")
plt.xlabel("Electric Vehicle Manufacturer")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```

```python
# selecting the top 3 mnaufacturer based on the number of vehicle registered


top_3_makes = ev_manufacturer_distribution.head(3).index

print(top_3_makes)



#filter the dataset for the top 3 makes

top_3_makes_data = ev_data[ev_data['Make'].isin(top_3_makes)]

print(top_3_makes_data)

#analyze the popularity of ev models within those top manufacturer

ev_model_distribution_top_makes = top_3_makes_data.groupby(['Make','Model']).size().sort_values(ascending = False)

print(ev_model_distribution_top_makes)

#top 10 models

top_models = ev_model_distribution_top_makes.head(10)

print(top_models)

plt.figure(figsize = (12,8))

top_models_df = top_models.reset_index().rename(columns={0: 'Number of Vehicles'})


sns.barplot(
    x= 'Model',
    y= 'Number of Vehicles',
    hue= 'Make',
    data = top_models_df,
    palette = "magma"
)

plt.title("Popularity of Electric Vehicle Models")
plt.xlabel("Electric Vehicle Model")
plt.ylabel("Number of Vehicles")
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()

```

```python
#ev Range analysis

ev_range_distribution = ev_data['Electric Range'].value_counts()

print(ev_range_distribution)

# how far can electric vehicle travel on single charge

plt.figure(figsize = (12,8))

sns.histplot(
    ev_data,
    bins = 30,
    x = 'Electric Range',
    kde = True,
    color = 'skyblue'
)

plt.title("Distribution of Electric Range")
plt.xlabel("Electric Range in miles")
plt.ylabel("Number of Vehicles")
plt.axvline(ev_data['Electric Range'].mean(), color = 'red', linestyle = 'dashed', linewidth = 2)
plt.tight_layout()
plt.show()
```

```python
#calculating average  electric range by model year

average_range_by_year = ev_data.groupby('Model Year')['Electric Range'].mean()

print(average_range_by_year)

plt.figure(figsize = (12,6))

sns.lineplot(
    x= 'Model Year',
    y= 'Electric Range',
    data = average_range_by_year.reset_index(),
    marker = 'o',
    color = 'green'
)

plt.title("Average Electric Range by Model Year")
plt.xlabel("Model Year")
plt.ylabel("Average Electric Range")
plt.grid(True)
plt.show()
```

```python
#how electric vehicle range vary among top manufacturer and models

top_makes_data_range = top_3_makes_data[['Make', 'Model', 'Electric Range']]

average_range_by_model = top_makes_data_range.groupby(['Make', 'Model'])['Electric Range'].mean().reset_index()

print(average_range_by_model)

top_range_model = average_range_by_model.head(10)

print(top_range_model)

plt.figure(figsize = (12,6))

sns.barplot(
    x = 'Model',
    y = 'Electric Range',
    hue = 'Make',
    data = top_range_model,
    palette= "viridis"
)

plt.title("Electric range variation among top manufacturer and  Models")
plt.xlabel("Electric Vehicle Manufacturer")
plt.ylabel("Average Electric Range")
plt.legend(title = 'Make')
plt.xticks(rotation = 45)
plt.tight_layout()
plt.show()
```
